```python
import numpy as np
```

```python
np.random.seed(23)
data = np.random.randint(60,100,200)
```

```python
def Calculate_Variance_Sample(data):
    n = len(data)
    
    if n <= 1:
        raise ValueError('Sample Size must be greater than 1.')
    sum_square = 0 
    
    average = np.mean(data)
    for i in range(n):
        s = data[i] - average 
        s_square = pow(s,2)
        sum_square += s_square
    
    return sum_square / (n - 1)
```

```python

def Calculate_Variance_Sample_2(data):
    n = len(data)
    if n <= 1:
        raise ValueError("Sample size must be greater than 1")
    
    average = np.mean(data)
    sum_square = 0
    
    for i in range(n):
        sum_square += (data[i] - average) ** 2
    
    return sum_square / (n - 1)
```

```python
def Calculate_Standard_Deviation_Sample(data):
    v = Calculate_Variance_Sample(data)
    return np.sqrt(v)
```

```python

def Calculate_Range(data):
    nn = np.min(data)
    ln = np.max(data)
    return ln - nn 
```

```python

variance_of_sample = Calculate_Variance_Sample(data)
print(f"The Variance of Sample Data: {variance_of_sample}")

dolechchuan = Calculate_Standard_Deviation_Sample(data)

print(f"The Standard Deviation of SD: {dolechchuan}")

print(f"The Range of DS: {Calculate_Range(data)}")
```

```python
def check_if_number(value):
    return isinstance(value, (int, float))

def Classify_Type(score):
    try:
        if not check_if_number(score) or score < 0 or score > 100:
            raise ValueError('Score is not valid!')

        if score >= 60 and score < 70:
            return 'Below Average'
        elif score >= 70 and score < 80: 
            return 'Satisfactory'
        elif score >= 80 and score < 90:
            return 'Good'
        elif score >= 90:
            return 'Excellent'

    except ValueError as e:
        print(f"Error: {e}")
        return None  
    
        
```
